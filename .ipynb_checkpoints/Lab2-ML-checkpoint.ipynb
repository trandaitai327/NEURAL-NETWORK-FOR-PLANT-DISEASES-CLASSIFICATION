# Tiền xử lý dữ liệu

- Kích cỡ mỗi `Batch`

```python
BATCH_SIZE = 32
```

- Tách tập train và tập validation

```python
X_train, X_val= train_test_split(df_train, test_size=0.3,random_state=0)
```

```python
X_train.head(5)
```

```python
X_val.head(5)
```

- `ImageDataGenerator` để tạo dữ liệu ở mỗi BATCH



```python
#Sử dụng cho tập train
train_datagen = ImageDataGenerator(rescale=1/255.0,
                                   rotation_range=5,
                                   zoom_range=0.1,
                                   shear_range=0.05,
                                   horizontal_flip=True,
                                   vertical_flip =True,
                                   width_shift_range=0.1,
                                   height_shift_range=0.1)
#Sử dụng cho tập test và validation
datagen = ImageDataGenerator(rescale=1/255.0)
```

- `flow_from_dataframe` để tạo DataFrame từ directory + tạo BATCH

```python
train_data = train_datagen.flow_from_dataframe(
    X_train,
    directory= 'images',
    x_col="image_id",
    y_col=["healthy","multiple_diseases","rust","scab"],
    color_mode="rgb",
    class_mode="raw",
    batch_size=BATCH_SIZE,
    shuffle=False)

val_data = datagen.flow_from_dataframe(
    X_val,
    directory= 'images',
    x_col="image_id",
    y_col=["healthy","multiple_diseases","rust","scab"],
    color_mode="rgb",
    class_mode="raw",
    batch_size=BATCH_SIZE,
    shuffle=False)

test_data = datagen.flow_from_dataframe(
    df_test,
    directory= 'images',
    x_col="image_id",
    color_mode="rgb",
    class_mode=None,
    batch_size=BATCH_SIZE,
    shuffle=False)
```

# Xây dựng mô hình

- Tạo `Base Model`

```python
input_shape = (256,256,3)
base_model_ds201 = DenseNet201(
       include_top=False,
       input_shape=input_shape,
       weights = 'imagenet')
base_model_itv3 = InceptionV3(
        include_top=False,
        input_shape=input_shape,
        weights = 'imagenet')
base_model_rn152v2 = ResNet152V2(
         include_top=False,
         input_shape=input_shape,
         weights = 'imagenet')
```

Lưu tất cả `base_model` vào `list_models`

```python
list_models = [(base_model_ds201,"DenseNet201"), (base_model_itv3,"InceptionV3"), (base_model_rn152v2,"ResNet152V2")]
```

- Tạo `Model tổng quát`

```python
"""MODEL Tổng quát"""
def Build_Model_CNN_Image(base_model ,nclasses,sparse_categorical, optimizor):
    model = Sequential()
    model.add(base_model)
    model.add(GlobalAveragePooling2D())
    model.add(Dense(1024, activation='relu'))
    model.add(BatchNormalization())
    
    model.add(Dense(nclasses,activation='softmax'))
    if sparse_categorical:
        model.compile(loss='sparse_categorical_crossentropy',
                      optimizer=optimizor,
                      metrics='accuracy')
    else:
        model.compile(loss='categorical_crossentropy',
                      optimizer=optimizor,
                      metrics='accuracy')
    return model
```

- Huấn luyện mô hình.

```python
models = [] #Lưu lại mô hình vừa huấn luyện
histories = [] #Lưu lại lịch sử huấn luyện mô hình.
for i, value in enumerate(list_models):
    print(value[1])
    model = Build_Model_CNN_Image(base_model=value[0],
                                  nclasses=4, 
                                  sparse_categorical=False, 
                                  optimizor="adam")
    filepath = "weights-best-file-{}.hdf5".format(value[1])
    #Lưu lại tham số mà val_accuracy lớn nhất
    checkpoint = ModelCheckpoint(filepath,
                                 monitor="val_accuracy",
                                 verbose=2,
                                 save_best_only=True)
    #Nếu val_accuracy không thay đổi thì dừng sớm.
    early_stopping = EarlyStopping(monitor='val_accuracy', 
                                   min_delta=0,
                                   patience=10, 
                                   verbose=2, 
                                   restore_best_weights=True)
    history = model.fit(x=train_data,
                        epochs=50,
                        validation_data=val_data,
                        steps_per_epoch=train_data.samples // BATCH_SIZE,
                        validation_steps=val_data.samples // BATCH_SIZE,
                        callbacks=[checkpoint, early_stopping],
                        verbose = 2)
    histories.append((value[1], history))
    models.append((value[1], model))
```

# Đánh giá mô hình và dự đoán

- Đánh giá lại mô hình với tập train và tập validation

```python
eval_train = []
eval_val = []
for i in range(len(models)):
    eval_train.append((models[i][0], models[i][1].evaluate(train_data)))
    eval_val.append((models[i][0], models[i][1].evaluate(val_data)))
```

```python
def create_df_loss_accuracy(eval):
    title = []
    loss = []
    accuracy = []
    for i in range(len(eval)):
        title.append(eval[i][0])
        loss.append(eval[i][1][0])
        accuracy.append(eval[i][1][1])
    return pd.DataFrame(data = {"Base model":title, "Loss":loss,"Accuracy":accuracy}).set_index("Base model")
```

Kết quả đánh giá với tập train

```python
create_df_loss_accuracy(eval_train)
```

Kết quả đánh giá với tập test

```python
create_df_loss_accuracy(eval_val)
```

- Vẽ biểu đồ lịch sử huấn luyện mô hình qua từng epoch

```python
def draw_acc_valacc_epochs(history, title):
    plt.plot(history.history['accuracy'])
    plt.plot(history.history['val_accuracy'])
    plt.title(title)
    plt.ylabel('accuracy')
    plt.xlabel('epoch')
    plt.legend(['train', 'validation'])
    plt.show()
```

```python
for i in range(len(histories)):
    draw_acc_valacc_epochs(histories[i][1], histories[i][0])
```

- Dự đoán từng mô hình huấn luyện với tập test

```python
y_pr_classes = {}
for i in range(len(models)):
    model = models[i][1]
    y_pr = model.predict(test_data)
    y_pr_classes[models[i][0]] =  np.argmax(y_pr, axis=-1)
```

```python
y_pr_df = pd.DataFrame(y_pr_classes)
y_pr_df
```

- Ba mô hình dự đoán giống nhau ở những test nào?

```python
condition = (y_pr_df["DenseNet201"] == y_pr_df["InceptionV3"])&(y_pr_df["DenseNet201"] == y_pr_df["ResNet152V2"])
temp_df = y_pr_df[condition]
temp_df
```

```python
print("Vậy ba mô hình dự đoán giống nhau {} chiếm {}".format(len(temp_df), len(temp_df)/ len(y_pr_df)))
```
